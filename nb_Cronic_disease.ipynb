```python
def map_gender(stratification):
    if stratification == 'Male':
        return 'Male'
    elif stratification == 'Female':
        return 'Female'
    else:
        return 'Other'
```

```python
df_disease['Gender'] = df_disease['Stratification1'].apply(map_gender)
```

```python
def map_race(stratification):
    races = ['American Indian or Alaska Native', 'Asian, non-Hispanic', 
             'White, non-Hispanic', 'Hispanic', 'Black, non-Hispanic', 
             'Asian or Pacific Islander', 'Multiracial, non-Hispanic', 
             'Other, non-Hispanic']
    if stratification in races:
        return stratification
    else:
        return 'Other'

```

```python
df_disease['Race'] = df_disease['Stratification1'].apply(map_race)
```

```python
# Extrae las coordenadas de la cadena
df_disease['GeoLocation'] = df_disease['GeoLocation'].str.strip('POINT ()')

# Divide la cadena en las coordenadas
df_disease[['Longitude', 'Latitude']] = df_disease['GeoLocation'].str.split(' ', expand=True)

# Convierte las nuevas columnas a números
df_disease['Long'] = pd.to_numeric(df_disease['Longitude'])
df_disease['Lat'] = pd.to_numeric(df_disease['Latitude'])

```

```python
df_disease = df_disease.drop(['Stratification1','GeoLocation','Longitude','Latitude'], axis=1)
```
